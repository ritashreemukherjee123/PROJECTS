Loading dataset CoNLL-2003

```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("juliangarratt/conll2003-dataset")


print(f"Contents of {path}:\n{os.listdir(path)}")

```

NER labels from CoNLL-2003

```python
ner_labels = [
    "O",
    "B-PER", "I-PER",
    "B-ORG", "I-ORG",
    "B-LOC", "I-LOC",
    "B-MISC", "I-MISC"
]
```

Add special labels


```python
special_labels = ["[CLS]", "[SEP]", "X", "[PAD]"]

#Merging all labels
all_labels = special_labels + ner_labels
```

Create tag2idx and idx2tag


```python
tag2idx = {tag: idx for idx, tag in enumerate(all_labels)}
idx2tag = {idx: tag for tag, idx in tag2idx.items()}
print("tag2idx:", tag2idx)
print("idx2tag:", idx2tag)
```

Using BERT Tokenizer


```python
from transformers import BertTokenizerFast

tokenizer = BertTokenizerFast.from_pretrained("bert-base-cased")
```

```python
words = ["EU", "rejects", "German", "call", "to", "boycott", "British", "lamb", "."]

encoding = tokenizer(
    words,
    is_split_into_words=True,
    truncation=True,
    padding="max_length",
    max_length=128,
    return_offsets_mapping=True
)
```

```python
labels = []
prev_word_id = None
word_labels = ["B-ORG", "O", "B-MISC", "O", "O", "O", "B-MISC", "O", "O"] # Sample labels for the 'words' in the previous cell

for i, word_id in enumerate(encoding.word_ids()):
    if word_id is None:
        labels.append(-100)
    elif word_id != prev_word_id:
        labels.append(tag2idx[word_labels[word_id]])
    else:
        labels.append(-100)
    prev_word_id = word_id
```
